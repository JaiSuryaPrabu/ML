# Importing Dataset

```python
import pandas as pd

data = pd.read_csv("house_price.csv")
data.head()
```

```python
print("Total number of rows",len(data))
print("Total number of missing values : ",data.isnull().sum())
```

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3)

print(f"Name\tSize\nX_train\t {len(x_train)}\nX_test\t {len(x_test)}\nY_train\t {len(y_train)}\nY_test\t {len(y_test)}\n ")
```

# Training Simple linear regression

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
prediction = model.predict(x_test)

#evaluation
from sklearn.metrics import r2_score
score = r2_score(y_test,prediction)
print("Score : ",score)
```

```python
print(f"The formula is : \n y = {model.coef_}*x + {model.intercept_}")
```

# Plotting

```python
import matplotlib.pyplot as plt

x = [i for i in range(1,33)]
plt.scatter(x,y_train)
plt.plot(x,model.predict(x_train),color="green")
plt.xlabel("Dataset")
plt.ylabel("Price")
plt.title("Training dataset")
```
